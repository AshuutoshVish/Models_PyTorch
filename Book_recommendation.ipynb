# Popularity Based Ratings

```python
ratings_with_name = ratings_df.merge(books_df,on='ISBN')
ratings_with_name
```

```python
ratings_with_name.shape
```

```python
ratings_with_name.groupby('Book-Title').count()['Book-Rating']
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df
```

```python
num_rating_df.rename(columns={'Book-Rating':'Ratings'},inplace=True)
num_rating_df
```

```python
users_ratings_df = ratings_df.merge(users_df, left_on = 'User-ID', right_on = 'User-ID')
users_ratings_df
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df.head(20)
```
